### Show 5 records from Netflix CSV file

```python
netflix_data = pd.read_csv('./datasets/netflix_data.csv')
netflix_data_df = pd.DataFrame(netflix_data)
netflix_data_df.head(5)
```

### Show 5 records and Scattered chart where type is movie with selected columns

```python
netflix_df = pd.DataFrame(netflix_data)
netflix_df_movies_only = netflix_df.where(netflix_df['type'] == 'Movie')
netflix_df_movies_selected_cols = netflix_df_movies_only[['title', 'country', 'genre', 'release_year', 'duration']]
netflix_five_movies = netflix_df_movies_selected_cols.head(5)
netflix_five_movies
```

```python
plt.scatter(netflix_five_movies['release_year'], netflix_five_movies['duration'])
plt.title('5 Neflix movie records')
```

### Show Movies under 60 mins duration

```python
short_movies = netflix_df_movies_only[netflix_df_movies_only['duration'] < 60]
short_movies = short_movies.head(20)
short_movies
```

### 8- Show non-featured films

```python
## Get colors genre wise.
colors = []
for index, row in short_movies.iterrows():
    if row['genre'] == 'Documentaries':
        colors.append('red')
    elif row['genre'] == 'Comedies':
        colors.append('blue')
    elif row['genre'] == 'Dramas':
        colors.append('green')
    else:
        colors.append('black')

## Show scattered chart with different colors (Genre wise)
plt.scatter(short_movies['release_year'], short_movies['duration'], c=colors)
plt.xlabel('Release Year')
plt.ylabel('Duration (min)')
plt.title('Short Netflix Movies by Genre')
plt.show()
```
