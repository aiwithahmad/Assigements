```python
import pandas as pd
import matplotlib.pyplot as plt
```

### 1. Display & Remove duplicate rows

```python
apps = pd.read_csv('./datasets/apps.csv');
apps_df = pd.DataFrame(apps);
sanitized_apps = apps_df.drop_duplicates();
total_apps = sanitized_apps.count()['App']
print('Total number of apps in the dataset = ', total_apps)
apps_df.head(50)
```

### 2. Sanitize Data

```python
special_chars = ['+', '$', '#', '%', '@', ';', ',']
cols_to_clean = ['Installs', 'Price']

def remove_special_char(val, strg):
    return val.replace(strg, '')

for col in cols_to_clean:
    for char in special_chars:
        #sanitized_apps[col] = apps[col].apply(remove_special_char(apps[col], char))
        sanitized_apps[col] = sanitized_apps[col].apply(lambda curr_val: remove_special_char(curr_val, char))
#        print(sanitized_apps[col])

sanitized_apps.info()
```

### 3. Correcting Data types

```python
print('Data type before modification: ', sanitized_apps['Installs'].dtype, sanitized_apps['Price'].dtype)
sanitized_apps['Installs'] = sanitized_apps['Installs'].astype(int)
sanitized_apps['Price'] = sanitized_apps['Price'].astype(float)
print('Data type after modification: ', sanitized_apps['Installs'].dtype, sanitized_apps['Price'].dtype)
```

### 4. Exploring App Categories

```python
sanitized_apps.head(50)
```

```python
## Get Unique categories count
categories = sanitized_apps['Category']
unique_cat = categories.unique()
print('Total categories: ', len(unique_cat), '\n')
category_wise_app_count = []
##  Get no: of apps in each category
for cat in unique_cat:
    curr_cat = sanitized_apps[sanitized_apps['Category'] == cat]
    category_wise_app_count.append((cat, curr_cat['App'].count()))
#    print(f"{cat}: {c['App'].count()} apps")

category_wise_app_count_df = pd.DataFrame(category_wise_app_count, columns=['Category', 'App_Count'])
category_wise_app_count_df['Category']
plt.bar(category_wise_app_count_df['Category'], category_wise_app_count_df['App_Count'])
plt.xlabel('Category')
plt.ylabel('Total no. of apps')
plt.title('Number of Apps by Category')
plt.xticks(rotation=90)
plt.show()
```

### 5 Distribution of app ratings

```python
ratings = sanitized_apps['Rating'].dropna()
avg_app_rating = sum(ratings) / len(ratings)
print('Average app rating = ', avg_app_rating)
```

```python
plt.hist(ratings, bins=30, edgecolor='orange', label='Avg app rating')
plt.axvline(avg_app_rating, color='red', linestyle='dashed', linewidth=1, label=f'Avg rating: {avg_app_rating:.2f}')
plt.legend()
plt.show()
```

### 6. Size & Price of an App.

```python
app_with_size_and_rating = sanitized_apps[sanitized_apps['Rating'].notnull() & sanitized_apps['Price'].notnull()]
large_cats = app_with_size_and_rating.groupby('Category').filter(lambda x: len(x) >= 250)
plt.scatter(large_cats['Size'], large_cats['Rating'], alpha=0.4, s=20)
plt.title("App Size vs. Rating (Categories with ≥250 Apps)")
plt.xlabel("Size (MB)")
plt.ylabel("Rating")
plt.show()
```
