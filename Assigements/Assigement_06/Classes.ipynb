### 1. Write a Python program to create a class representing a Circle. Include methods to calculate its area and perimeter.

```python
import math
class Circle:
    #Constructor
    def __init__(self,Radius):
        self.Radius = Radius
        
    # Method
    def Area(self,Radius):
        Area = math.pi*Radius**2
        return Area

    def perimeter(self, Radius):
        perimeter = 2*math.pi*Radius
        return perimeter

circle_1 =Circle(5)
Area = circle_1.Area(5)
print("The Area is = ",round(Area,2))

perimeter = circle_1.perimeter(5)
print("The Perimeter is = ", round(perimeter,2))
```

### 2. Write a Python program to create a person class. Include attributes like name, country and date of birth. Implement a method to determine the person's age.\n

```python
from datetime import date

class Person:
    def __init__(self, name, country, dob):
        self.name = name
        self.country =country
        self.dob = dob
        
    def Agecalculate(self):
        today = date.today()
        age = today.year - self.dob.year
        return age

man = Person("Qasim", "Pakistan" ,date(2001,5,7))
personage = man.Agecalculate()
print(personage)
```

### 3. Write a Python program to create a calculator class. Include methods for basic arithmetic operations.

```python
class Calculator:
    def __init__ (self, operator, num1, num2):
        self.operator = operator
        self.num1 =num1
        self.num2 = num2

    def Calculate(self):
        if self.operator == "+":
            ans = self.num1 + self.num2
        elif self.operator == "-":
            ans = self.num1 - self.num2
        elif self.operator == "*":
            ans = self.num1 * self.num2
        elif self.operator == "/":
            ans = self.num1 / self.num2
        elif self.operator == "**":
            ans = self.num1 ** self.num2
        return ans

calculation_1 = Calculator("*", 2,2)
print(calculation_1.Calculate())
```
